```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
def plot_main_distribution(plot_dots=True):
    lines = list(map(str.split, open("data/1_theor_dens.txt").readlines()))
    v = lines[0][0]
    x = [float(elem[0]) for elem in lines[1:]]
    y = [float(elem[1]) for elem in lines[1:]]
    plt.plot(x, y, color='cyan', label='Теоретическая плотность')
    
    lines = list(map(str.split, open("data/1_emperical_dens.txt").readlines()))
    x = [float(elem[0]) for elem in lines]
    y = [float(elem[1]) for elem in lines]
    k = len(x) // 2
    plt.plot(x, y, color='#6FD890', label='Эмпирическая плотность')
    
    lines = list(map(str.split, open("data/1_generated_dots.txt").readlines()))
    x = [float(elem[0]) for elem in lines]
    y1 = [float(elem[1]) for elem in lines]
    y2 = [float(elem[2]) for elem in lines]
    N = len(x)
    
    if plot_dots:
        plt.plot(x, y1, 'bo', alpha=0.7, markersize=3)
        plt.plot(x, y2, 'go', alpha=0.7, markersize=3)
    
    plt.legend()
    plt.title(f'Основное распределение\nv={v}, N={N}, k={k}');
    
def plot_shift_scaled_distribution(plot_dots=True):
    lines = list(map(str.split, open("data/2_theor_dens.txt").readlines()))
    v, mu, lambd = lines[0][0], lines[0][1], lines[0][2]
    x = [float(elem[0]) for elem in lines[1:]]
    y = [float(elem[1]) for elem in lines[1:]]
    plt.plot(x, y, color='cyan', label='Теоретическая плотность')
    
    lines = list(map(str.split, open("data/2_emperical_dens.txt").readlines()))
    x = [float(elem[0]) for elem in lines]
    y = [float(elem[1]) for elem in lines]
    k = len(x) // 2
    plt.plot(x, y, color='#6FD890', label='Эмпирическая плотность')
    
    lines = list(map(str.split, open("data/2_generated_dots.txt").readlines()))
    x = [float(elem[0]) for elem in lines]
    y1 = [float(elem[1]) for elem in lines]
    y2 = [float(elem[2]) for elem in lines]
    N = len(x)
    
    if plot_dots:
        plt.plot(x, y1, 'bo', alpha=0.7, markersize=3)
        plt.plot(x, y2, 'go', alpha=0.7, markersize=3)
    
    plt.legend()
    plt.grid(linestyle=':')
    plt.title(f'Распределение после преобразования сдвига-растяжения\nv={v}, μ={mu}, λ={lambd}, N={N}, k={k}')
    
def plot_mixture_distribution(plot_dots=True):
    lines = list(map(str.split, open("data/3_theor_dens.txt").readlines()))
    v1, mu1, lambda1, v2, mu2, lambda2, p = lines[0][0], lines[0][1], lines[0][2], lines[0][3], lines[0][4], lines[0][5], lines[0][6]
    x = [float(elem[0]) for elem in lines[1:]]
    y = [float(elem[1]) for elem in lines[1:]]
    plt.plot(x, y, color='cyan', label='Теоретическая плотность')
    
    lines = list(map(str.split, open("data/3_emperical_dens.txt").readlines()))
    x = [float(elem[0]) for elem in lines]
    y = [float(elem[1]) for elem in lines]
    k = len(x) // 2
    plt.plot(x, y, color='#6FD890', label='Эмпирическая плотность')
    
    lines = list(map(str.split, open("data/3_generated_dots.txt").readlines()))
    x = [float(elem[0]) for elem in lines]
    y1 = [float(elem[1]) for elem in lines]
    y2 = [float(elem[2]) for elem in lines]
    N = len(x)
    
    if plot_dots:
        plt.plot(x, y1, 'bo', alpha=0.7, markersize=3)
        plt.plot(x, y2, 'go', alpha=0.7, markersize=3)
    
    plt.legend()
    plt.grid(linestyle=':')
    plt.title(f'Распределение смеси\nv1={v1}, μ1={mu1}, λ1={lambda1}, v2={v2}, μ2={mu2}, λ2={lambda2}, p={p} N={N}, k={k}')
    
def plot_emperical_from_emperical_distribution(plot_dots=True):
    lines = list(map(str.split, open("data/4_theor_dens.txt").readlines()))
    v, mu, lambd = lines[0][0], lines[0][1], lines[0][2]
    x = [float(elem[0]) for elem in lines[1:]]
    y = [float(elem[1]) for elem in lines[1:]]
    plt.plot(x, y, color='cyan', label='Теоретическая плотность')
    
    lines = list(map(str.split, open("data/4_emperical_dens.txt").readlines()))
    x = [float(elem[0]) for elem in lines]
    y = [float(elem[1]) for elem in lines]
    plt.plot(x, y, color='#6FD890', label='Эмпирическая плотность')
    
    lines = list(map(str.split, open("data/4_emperical_dens2.txt").readlines()))
    x = [float(elem[0]) for elem in lines]
    y = [float(elem[1]) for elem in lines]
    k = len(x) // 2
    plt.plot(x, y, color='#FF5733', label='Эмпирическая плотность на основе\n сгенерированных по эмперическому распределению')
    
    lines = list(map(str.split, open("data/4_generated_dots.txt").readlines()))
    x = [float(elem[0]) for elem in lines]
    y1 = [float(elem[1]) for elem in lines]
    y2 = [float(elem[2]) for elem in lines]
    y3 = [float(elem[3]) for elem in lines]
    N = len(x)
    
    if plot_dots:
        plt.plot(x, y1, 'bo', alpha=0.7, markersize=3)
        plt.plot(x, y2, 'go', alpha=0.7, markersize=3)
        plt.plot(x, y3, 'ro', alpha=0.7, markersize=3)
    
    
    plt.legend(bbox_to_anchor=(1.05, 1))
    plt.grid(linestyle=':')
    plt.title(f'Распределение на основе эмпирического распределения\nv={v}, μ={mu}, λ={lambd}, N={N}, k={k}')
```
