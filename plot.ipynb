```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
def plot_using_hist():
    file = open('data/2_1.txt', encoding='utf-8')
    column = pd.Series(list(map(float, file.readlines())))
    column.hist(bins=100, density=True, alpha=0.7)

    file = open('data/1_1.txt', encoding='utf-8')
    data = np.array(list(map(float, file.readlines())))
    plt.plot(np.arange(-0.99, 1, 0.01), data, color='red');
    
def plot_using_density():
    file = open('data/1_1.txt', encoding='utf-8')
    data = np.array(list(map(float, file.readlines())))
    plt.plot(np.arange(-0.99, 1, 0.01), data, color='red');
    
    file = open('data/3_1.txt', encoding='utf-8')
    x, y = [], []
    for row in file:
        values = list(map(float, row.split()))
        x.append(values[0]), x.append(values[1])
        y.append(values[2]), y.append(values[2])
    plt.plot(x, y, alpha=0.7)
    
def plot_mixture_data():
    
    file = open('data/4_1.txt', encoding='utf-8')
    column = pd.Series(list(map(float, file.readlines())))
    column.hist(density=True, bins=100, alpha=0.7)
    
    file = open('data/4_2.txt', encoding='utf-8')
    x, y = [], []
    for row in file:
        values = list(map(float, row.split()))
        x.append(values[0]), x.append(values[1])
        y.append(values[2]), y.append(values[2])
    plt.plot(x, y, alpha=0.7)
    
def plot_emperical_generated_data():
    # Гистограмма для случайных величин
    file = open('data/5_1.txt', encoding='utf-8')
    column = pd.Series(list(map(float, file.readlines())))
    column.hist(bins=100, density=True, alpha=0.7)
    
    # Гистограмма для случайных величин из распределения эмпирической плотности
    file = open('data/5_2.txt', encoding='utf-8')
    column = pd.Series(list(map(float, file.readlines())))
    column.hist(density=True, bins=100, alpha=0.3, color='green')
    
    # Эмпирическая плотность
    file = open('data/5_3.txt', encoding='utf-8')
    x, y = [], []
    for row in file:
        values = list(map(float, row.split()))
        x.append(values[0]), x.append(values[1])
        y.append(values[2]), y.append(values[2])
    plt.plot(x, y, alpha=0.7, color='black')
    
    # Теоретическая плотность shift-scaled
    file = open('data/5_4.txt', encoding='utf-8')
    x, y = [], []
    for row in file:
        values = list(map(float, row.split()))
        x.append(values[0]), y.append(values[1])
    plt.plot(x, y, color='red')
```

```python
plot_using_hist()
plt.title('Теоретическая плотность основного распределения и соответствующие сгенерированные случайные величины');
```

```python
plot_using_density()
plt.title('Теоретическая плотность основного распределения и эмпирическая плотность');
```

```python
plot_mixture_data()
plt.title('Распределение смеси и эмпирическая плотность');
```

```python
plot_emperical_generated_data()
plt.title('Генерация случайных величин на основе эмпирической плотности');
```
